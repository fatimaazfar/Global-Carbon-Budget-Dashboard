# Feature Engineering for DF1

```python
merged_df = Global_Carbon_Budget.merge(Historical_Budget, on="Year", how="outer")
merged_df = merged_df.merge(Fossil_Emissions_by_Category, on="Year", how="outer")
merged_df = merged_df.merge(Ocean_Sink, on="Year", how="outer")
merged_df = merged_df.merge(Terrestrial_Sink, on="Year", how="outer")
merged_df = merged_df.merge(Cement_Carbonation_Sink, on="Year", how="outer")
merged_df
```

```python
merged_df.describe()
```

```python
merged_df.columns
```

```python
import numpy as np

# averaging the duplicates
merged_df['fossil emissions excluding carbonation'] = merged_df[['fossil emissions excluding carbonation_x', 'fossil emissions excluding carbonation_y', 'fossil.emissions.excluding.carbonation']].mean(axis=1)
merged_df['land_sink'] = merged_df[['land sink_x', 'land sink_y']].mean(axis=1)
merged_df['Multi-model mean'] = merged_df[['Multi-model mean_x', 'Multi-model mean_y']].mean(axis=1)
merged_df['GCB'] = merged_df[['GCB_x', 'GCB_y']].mean(axis=1)

# dropping the original columns
merged_df.drop(['fossil emissions excluding carbonation_x', 'fossil emissions excluding carbonation_y', 'land sink_x', 'land sink_y'], axis=1, inplace=True)

# total emissions = add fossil emissions (excluding carbonation), land-use change emissions, and cement carbonation sink
merged_df['total_emissions'] = merged_df['fossil emissions excluding carbonation'] + merged_df['land-use change emissions'] + merged_df['cement carbonation sink']

# total carbon sinks = adding ocean sink, land sink, and cement carbonation sink
merged_df['total_sinks'] = merged_df['ocean sink'] + merged_df['land_sink'] + merged_df['cement carbonation sink']

# emissions to sinks ratio = total emissions/total sinks, this features is used to understand the balance between emissions and nature's ability to absorb carbon
merged_df['emissions_to_sinks_ratio'] = merged_df['total_emissions'] / merged_df['total_sinks']

# proportion of each fossil fuel type in total fossil emissions
merged_df['coal_proportion'] = merged_df['Coal'] / merged_df['fossil emissions excluding carbonation']
merged_df['oil_proportion'] = merged_df['Oil'] / merged_df['fossil emissions excluding carbonation']
merged_df['gas_proportion'] = merged_df['Gas'] / merged_df['fossil emissions excluding carbonation']

# Normalized emissions and sinks to understand their relative magnitudes on a comparable scale
columns_to_normalize = ['fossil emissions excluding carbonation', 'land-use change emissions', 'ocean sink', 'land_sink', 'cement carbonation sink']
df_normalized = merged_df[columns_to_normalize].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

merged_df = pd.concat([merged_df, df_normalized.add_suffix('_normalized')], axis=1)

# Yearly change in total emissions and sinks to track how they evolve over time
merged_df = merged_df.sort_values(by='Year')
merged_df['yearly_change_in_emissions'] = merged_df['total_emissions'].diff()

# change in atmospheric growth year-over-year to understand acceleration or deceleration in growth
merged_df['atmospheric_growth_change'] = merged_df['atmospheric growth'].diff()

# fossil fuel diversity index as a measure of reliance on different fuel types
# a higher value indicates a more balanced mix of fossil fuel sources.
merged_df['fossil_fuel_diversity'] = 1 / ((merged_df['coal_proportion'] ** 2) + (merged_df['oil_proportion'] ** 2) + (merged_df['gas_proportion'] ** 2))

# carbon intensity of energy, assuming 'fossil emissions excluding carbonation' as a proxy for total fossil fuel energy
# this feature represents the amount of emissions per unit of energy derived from fossil fuels
merged_df['carbon_intensity'] = merged_df['fossil emissions excluding carbonation'] / (merged_df['Coal'] + merged_df['Oil'] + merged_df['Gas'] + merged_df['Other'])

# normalize new features to enable comparison across years and different magnitudes
features_to_normalize = ['atmospheric_growth_change', 'fossil_fuel_diversity', 'carbon_intensity']
df_normalized_new = merged_df[features_to_normalize].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
merged_df = pd.concat([merged_df, df_normalized_new.add_suffix('_normalized')], axis=1)

merged_df.sort_values(by='Year', inplace=True)
```

# Feature Engineering for DF2

```python
merged_df2 = pd.merge(Global_Carbon_Budget_df2, Territorial_Emissions, on='Year', how='outer')
merged_df2 = pd.merge(merged_df2, Consumption_Emissions, on='Year', how='outer')
merged_df2 = pd.merge(merged_df2, Emissions_Transfers, on='Year', how='outer')

nan_percentage = merged_df2.isna().mean() * 100
columns_to_drop = nan_percentage[nan_percentage > 80].index
merged_df2 = merged_df2.drop(columns=columns_to_drop)
merged_df2  = merged_df2.dropna()

merged_df2
```

```python
merged_df2.columns
```

```python
merged_df2.columns = merged_df2.columns.str.replace('_x$', '', regex=True)
unique_countries_regions = set(col.split('_')[0] for col in merged_df2.columns if '_' in col)

for country_region in unique_countries_regions:
    if f'{country_region}_y' in merged_df2.columns:
        merged_df2[country_region] = merged_df2[[f'{country_region}_x', f'{country_region}_y']].mean(axis=1)
        merged_df2.drop([f'{country_region}_x', f'{country_region}_y'], axis=1, inplace=True)
    else:
        merged_df2.rename({f'{country_region}_x': country_region}, axis=1, inplace=True)

categories = ['KP Annex B', 'Non KP Annex B', 'OECD', 'Non-OECD', 'EU27', 'Africa', 'Asia', 'Central America', 'Europe', 'Middle East', 'North America', 'Oceania', 'South America', 'Bunkers', 'Statistical Difference', 'World']
for category in categories:
    merged_df2.rename({f'{category}_x': category}, axis=1, inplace=True)

merged_df2
```

```python
# Total Global Emissions
merged_df2['total_global_emissions'] = merged_df2['fossil emissions excluding carbonation'] + merged_df2['land-use change emissions'] + merged_df2['cement carbonation sink']

# Emissions Growth Rate
merged_df2['emissions_growth_rate'] = merged_df2['total_global_emissions'].diff()

# Sink Efficiency
merged_df2['sink_efficiency'] = (merged_df2['ocean sink'] + merged_df2['land sink'] + merged_df2['cement carbonation sink']) / merged_df2['total_global_emissions']

# Regional Emissions OECD
oecd_countries = [col for col in merged_df2.columns if 'OECD' in col and '_' not in col]
merged_df2['regional_emissions_OECD'] = merged_df2[oecd_countries].sum(axis=1)

# Budget Imbalance Ratio
merged_df2['budget_imbalance_ratio'] = merged_df2['budget imbalance'] / merged_df2['total_global_emissions']

# Cumulative Emissions
merged_df2['cumulative_emissions'] = merged_df2['total_global_emissions'].cumsum()

merged_df2
```

```python
merged_df2 = merged_df2.dropna()
```
